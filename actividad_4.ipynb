```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = r"D:\Data-Science\Poblaci_n_Antioquia_censada_2018_por_tipo_territorio__sexo_y_edades_simples_20240330.csv"
```

```python
datos = pd.read_csv(dataset, delimiter=';', encoding='latin1')
```

```python
# 1. Abstracto con motivación y audiencia
"""
El objetivo de este análisis es comprender la distribución demográfica en diferentes municipios y áreas (cabecera, centro poblado y rural disperso) en función de la edad y el género. Este análisis será útil para autoridades gubernamentales, investigadores y organizaciones interesadas en políticas públicas, planificación urbana y desarrollo comunitario.
"""

# 2. Preguntas/Hipótesis que queremos resolver mediante el análisis de datos
"""
¿Cuál es la distribución de la población por edad y género en los diferentes tipos de áreas dentro de los municipios?
¿Existen diferencias significativas en la distribución demográfica entre las áreas de cabecera, centro poblado y rural disperso?
¿Hay alguna relación entre la edad y el género en las diferentes áreas?
"""
```

```python
print(datos.head())
```

```python
summary_stats = datos.describe()
print(summary_stats)
```

```python
sns.histplot(datos['Edad'], bins=20)
plt.title('Distribución de Edad')
plt.show()
```

```python
sns.boxplot(x='Hombres_Cabecera', y='Edad', data=datos)
plt.title('Distribución de Edad por Hombres en Cabecera')
plt.show()
```

```python
sns.boxplot(x='Mujeres_Cabecera', y='Edad', data=datos)
plt.title('Distribución de Edad por Mujeres en Cabecera')
plt.show()
```
