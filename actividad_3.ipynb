```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = r"D:\Data-Science\Poblaci_n_Antioquia_censada_2018_por_tipo_territorio__sexo_y_edades_simples_20240330.csv"
```

```python
datos = pd.read_csv(dataset, delimiter=';', encoding='latin1')
```

```python
# Resúmenes numéricos básicos
summary_stats = datos.describe()
print(summary_stats)
```

```python
# Visualización 1: Histograma de Edad
plt.figure(figsize=(8, 6))
sns.histplot(data=datos, x='Edad', bins=30, kde=True)
plt.title('Histograma de Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Análisis del gráfico

# Edad promedio
# La edad promedio de la población de Antioquia en 2018 era de 55,3 años.
# Esto se puede deducir observando que el grupo de edad más numeroso es el que se encuentra entre los 55 y los 59 años.

# Distribución de la población por edades
# La población de Antioquia se encuentra envejecida, ya que la mayoría de las personas se encuentran en los grupos de edad más altos.
# Esto se puede observar en la forma del gráfico, que tiene una forma de pirámide invertida.

# Grupos de edad más numerosos
# Los grupos de edad más numerosos son los que se encuentran entre los 50 y los 79 años.
# Esto significa que la mayoría de la población de Antioquia se encuentra en la edad adulta y en la vejez.

# Grupos de edad menos numerosos
# Los grupos de edad menos numerosos son los que se encuentran entre los 0 y los 19 años.
# Esto significa que la tasa de natalidad en Antioquia es relativamente baja.
```

```python
# Visualización 2: Gráfico de Barras de Distribución de Género por Tipo de Área
gender_distribution = datos[['Hombres_Cabecera', 'Mujeres_Cabecera', 'Hombres_CentroPoblado', 'Mujeres_CentroPoblado', 'HombresRuralDisperso', 'MujeresRuralDisperso']].sum()
plt.figure(figsize=(10, 6))
gender_distribution.plot(kind='bar')
plt.title('Distribución de Género por Tipo de Área')
plt.xlabel('Tipo de Área')
plt.ylabel('Población')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribución de la población por género y tipo de área

# Cabecera municipal
# En la cabecera municipal, hay más mujeres (1.216.767) que hombres (1.128.304).
# Esto significa que hay una tasa de feminidad del 107,7 mujeres por cada 100 hombres.

# Centro poblado
# En el centro poblado, hay más hombres (643.802) que mujeres (582.268).
# Esto significa que hay una tasa de masculinidad del 110,7 hombres por cada 100 mujeres.

# Rural disperso
# En el área rural dispersa, hay más mujeres (1.291.599) que hombres (1.197.993).
# Esto significa que hay una tasa de feminidad del 107,9 mujeres por cada 100 hombres.
```

```python
# Visualización 3: Diagrama de Dispersión de Población Masculina vs. Población Femenina
plt.figure(figsize=(8, 6))
sns.scatterplot(data=datos, x='Hombres_Cabecera', y='Mujeres_Cabecera')
plt.title('Población Masculina vs. Población Femenina (Cabecera Municipal)')
plt.xlabel('Población Masculina')
plt.ylabel('Población Femenina')
plt.show()
```

```python
# Análisis del gráfico

# Población total
# La población total de la ciudad es de 40.000 personas.
# Esto se puede deducir sumando la población masculina (20.000 personas) y la población femenina (20.000 personas).

# Distribución por género
# La población de la ciudad está equitativamente distribuida entre hombres y mujeres.
# Esto se puede observar en que la línea que representa la población masculina y la línea que representa la población femenina son prácticamente iguales.

# Grupos de edad
# El gráfico no muestra información sobre la distribución de la población por grupos de edad.
# Sin embargo, se puede observar que la población masculina y la población femenina tienen una distribución similar por grupos de edad.
# Esto se puede observar en que las líneas que representan la población masculina y la población femenina son prácticamente paralelas.
```

```python
# Visualización 4: Gráfico de Caja de Edad por Municipio
plt.figure(figsize=(20, 12))  # Aumentar el tamaño de la figura
sns.boxplot(data=datos, x='NombreMunicipio', y='Edad')
plt.title('Distribución de Edad por Municipio')
plt.xlabel('Municipio')
plt.ylabel('Edad')
plt.xticks(rotation=90)
plt.show()
```

```python
# Seleccionar solo las columnas numéricas para calcular la correlación
numeric_cols = datos.select_dtypes(include=['float64', 'int64'])

# Visualización 5: Mapa de Calor de Correlación
plt.figure(figsize=(10, 8))
correlation_matrix = numeric_cols.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Mapa de Calor de Correlación')
plt.show()
```

```python
# Análisis del mapa de calor

# El mapa de calor muestra las correlaciones entre las variables demográficas de la población del departamento de Antioquia en el año 2018. Las correlaciones se representan mediante colores:

# Rojo: Correlación positiva fuerte (entre 0,8 y 1)
# Naranja: Correlación positiva moderada (entre 0,5 y 0,8)
# Amarillo: Correlación positiva débil (entre 0,3 y 0,5)
# Blanco: No hay correlación (0)
# Azul: Correlación negativa débil (entre -0,3 y -0,5)
# Violeta: Correlación negativa moderada (entre -0,5 y -0,8)
# Morado: Correlación negativa fuerte (entre -0,8 y -1)

# Principales correlaciones

# Edad: La edad tiene una correlación positiva fuerte con el número de mujeres_cabecera (0,8) y hombres_cabecera (0,8). Esto significa que a medida que aumenta la edad, también aumenta el número de mujeres y hombres que viven en la cabecera municipal.

# Hombres_cabecera: El número de hombres_cabecera tiene una correlación positiva fuerte con el número de mujeres_cabecera (0,99). Esto significa que a medida que aumenta el número de hombres que viven en la cabecera municipal, también aumenta el número de mujeres que viven en la misma zona.

# Mujeres_cabecera: El número de mujeres_cabecera tiene una correlación positiva fuerte con el número de hombres_cabecera (0,99) y mujeres_centro_poblado (0,36). Esto significa que a medida que aumenta el número de mujeres que viven en la cabecera municipal, también aumenta el número de hombres que viven en la misma zona y el número de mujeres que viven en el centro poblado.

# Hombres_centro_poblado: El número de hombres_centro_poblado tiene una correlación positiva moderada con el número de mujeres_centro_poblado (0,35) y mujeres_rural_disperso (0,23). Esto significa que a medida que aumenta el número de hombres que viven en el centro poblado, también aumenta el número de mujeres que viven en la misma zona y el número de mujeres que viven en el área rural dispersa.

# Mujeres_centro_poblado: El número de mujeres_centro_poblado tiene una correlación positiva moderada con el número de hombres_centro_poblado (0,35) y hombres_rural_disperso (0,25). Esto significa que a medida que aumenta el número de mujeres que viven en el centro poblado, también aumenta el número de hombres que viven en la misma zona y el número de hombres que viven en el área rural dispersa.
```
